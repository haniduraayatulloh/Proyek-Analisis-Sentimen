# Mengimpor Library



```python

import pandas as pd
from google_play_scraper import app, reviews, Sort
```

# Mengambil Data dengan API



```python

app_info = app('com.zhiliaoapp.musically', lang='id', country='id')
print("App Info:")
print(app_info)


result, continuation_token = reviews(
    'com.zhiliaoapp.musically',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=10000  
)

print(f"Retrieved {len(result)} reviews")
```

# Mengurai dan Memproses Data



```python

df = pd.DataFrame(result)
print("Columns:", df.columns.tolist())
print(df.head())
```

# Menarik Data Penting



```python

important_columns = ['reviewId', 'userName', 'userImage', 'content', 'score', 'thumbsUpCount', 'reviewCreatedVersion', 'at', 'replyContent', 'repliedAt']
df_selected = df[important_columns]
print(df_selected.head())
```

# Menyimpan Data ke File



```python

df_selected.to_csv('tiktok_google_play_reviews.csv', index=False)
print("Data saved to tiktok_google_play_reviews.csv")
```
